```python
def encode_categorical_column(df, input_col, output_col):
   encoder = LabelEncoder()
   df[output_col] = encoder.fit_transform(df[input_col])
   
   mapping = dict(zip(encoder.classes_, encoder.transform(encoder.classes_)))
   
   with open('Use_mapping.json', 'w') as f:
       json.dump(mapping, f)
       
   return df

# Usage:
df2 = encode_categorical_column(df, 'Harmonized Functional Use', 'Harmonized Functional Use Encoded')
```

```python
import json
import numpy as np
from sklearn.preprocessing import LabelEncoder

def encode_categorical_column(df, input_col, output_col):
   encoder = LabelEncoder()
   df[output_col] = encoder.fit_transform(df[input_col])
   
   # Convert numpy int64 to regular int for JSON serialization
   mapping = {k: int(v) for k, v in 
             zip(encoder.classes_, encoder.transform(encoder.classes_))}
   
   with open(f'{input_col}_mapping.json', 'w') as f:
       json.dump(mapping, f)
   
   return df

df2 = encode_categorical_column(df, 'Harmonized Functional Use', 'Harmonized Functional Use Encoded')
```
