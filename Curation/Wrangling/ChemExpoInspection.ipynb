```python
import pandas as pd
import matplotlib.pyplot as plt

def plot_and_print_unique_counts(dataframe):
    unique_counts = dataframe.nunique()
    
    # Print the unique counts
    print("Number of unique entries in each column:")
    print(unique_counts)
    print("-" * 40)
    
    # Plot the unique counts
    plt.figure(figsize=(10, 6))
    unique_counts.plot(kind='bar')
    plt.title('Number of Unique Entries in Each Column')
    plt.xlabel('Columns')
    plt.ylabel('Unique Count')
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()

# Example usage
# df = pd.DataFrame({
#     'A': [1, 2, 3, 1],
#     'B': ['apple', 'orange', 'banana', 'apple'],
#     'C': [True, False, True, False]
# })

# plot_and_print_unique_counts(df)
```
