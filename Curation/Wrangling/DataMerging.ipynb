```python
import pandas as pd
import numpy as np
```

```python

# read in features
Desc = pd.read_csv('RightSmilesFeaturized.csv')
Desc = Desc.dropna()
```

```python
# read in use case data

data = pd.read_csv('ReducedUseCases.csv')
data = data.drop(['Unnamed: 0'], axis =1)
```

```python
data = data.rename(columns={'Curated CAS':'CAS'})
```

```python
data
```

```python
# merge the two

merged = data.merge(Desc, on='CAS',how='left')
```

```python
merged
```

```python
merged = merged.dropna(subset=['SMILES','Harmonized Functional Use'])
```

```python
merged = merged.drop(['ROMol'],axis=1)
```

```python
merged
```

```python
merged.to_csv('../../Data/Curated/UseCaseDataCurated.csv',index=False)
```
